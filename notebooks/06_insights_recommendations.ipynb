### Cost Savings Analysis

```
cost_per_readmission = 15000
flagged_patients_per_year = 500
prevented_rate = 0.20

prevented_count = flagged_patients_per_year * prevented_rate
annual_savings = prevented_count * cost_per_readmission

print(f'Estimated Readmissions Prevented per Year: {int(prevented_count)}')
print(f'Estimated Annual Savings: ${annual_savings:,.2f}')
```
