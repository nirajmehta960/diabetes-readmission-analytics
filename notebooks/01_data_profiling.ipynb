# Notebook 01: Data Profiling
Initial exploration and quality assessment of the raw diabetic dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

if os.path.basename(os.getcwd()) == "notebooks":
    os.chdir("..")

df = pd.read_csv("data/raw/diabetic_data.csv")
print(f"Data Shape: {df.shape}")
```

### Missing Values

```
missing = df.isnull().sum()
missing = missing[missing > 0]
if missing.empty:
    print("No explicitly missing values (NaN), but let's check for '?'")

# Replacing '?' with NaN to check actual missingness
df.replace("?", pd.NA, inplace=True)
missing_actual = df.isnull().sum()
missing_actual = missing_actual[missing_actual > 0]
print(missing_actual.sort_values(ascending=False))
```

### Basic Distributions

```
sns.countplot(data=df, x="readmitted")
plt.title("Distribution of Output Label (readmitted)")
plt.show()
```
